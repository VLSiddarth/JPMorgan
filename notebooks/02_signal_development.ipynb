# 02 - Signal Development Notebook
This notebook demonstrates development and testing of momentum, value, and macro signals.

```
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

## Load Sample Data

```
dates = pd.date_range(end=datetime.now(), periods=200)
prices = pd.Series(np.random.normal(0,1,200).cumsum()+100, index=dates)
prices.head()
```

## Momentum Signal Example

```
momentum = prices.pct_change(20)
momentum.tail()
```

## Value Signal Placeholder

```
pe_ratio = pd.Series(np.random.uniform(8,20,200), index=dates)
value_signal = 1/pe_ratio
value_signal.tail()
```

## Macro Signal Placeholder

```
macro_index = pd.Series(np.random.normal(0,1,200).cumsum(), index=dates)
macro_index.tail()
```
